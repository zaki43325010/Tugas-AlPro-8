#Praktikum 01 – Fungsi Dasar dengan def

```python
def sapa():
  print("Halo, Selamat datang di Politeknik Negeri Semarang!")

sapa()
```

#Praktikum 02 – Fungsi dengan Parameter dan Return

```python
def tambah(a,b):
  return a+b

hasil=tambah(7, 3)
print("Hasil penjumlahan: ", hasil)
```

#Praktikum 03 – Fungsi dengan Default Argument

```python
def sapa(nama="Mahasiswa"):
  print(f"Halo, {nama}")
sapa()
sapa("Andi")
```

#Praktikum 04 – Fungsi dengan *args dan **kwargs

```python
def jumlahkan(*args):
  return sum(args)
def biodata(**kwargs):
  for key, value in kwargs.items():
    print(f"{key}:{value}")
print("Total:", jumlahkan(1,2,3,4,5))
biodata(Nama="Budi",Umur=20,Prodi="TRK")
```

#Praktikum 05 – Scope Variabel

```python
x=10
def ubah():
  x=5
  print("Nilai lokal:",x)
ubah()
print("Nilai global:",x)
```

#Praktikum 07 – Fungsi Decorator

```python
def decorator_sapa(fungsi):
  def wrapper():
    print("Sebelum fungsi dijalankan...")
    fungsi()
    print("Setelah fungsi dijalankan...")
  return wrapper
@decorator_sapa
def sapa():
  print("Halo, Mahasiswa Polines!")

sapa()
```

#Praktikum 08 – Studi Kasus: Menghitung Rata-rata Nilai Mahasiswa

```python
def rata(data):
  return sum(data)/len(data)

nilai=[80,90,85]
rata_rata=rata(nilai)
print("Rata-rata nilai:", rata_rata)
```

#1.	Fungsi Sederhana

```python
def perkenalan(nama:"Mahasiswa",prodi:"Prodi"):
  print(f"Halo, nama saya {nama}, saya dari {prodi}")

perkenalan("Andi", "Teknik Listrik")
perkenalan("Asep", "Teknik Mesin")
perkenalan("Ucok", "Teknik Informatika")
```

#2.	Fungsi dengan Parameter dan Return

```python
def luas_persegi(panjang, lebar):
  return panjang*lebar

hasil=luas_persegi(7, 8)
print("Hasil luas persegi:", hasil)
```

#3.	Fungsi dengan Default Argument

```python
def selamat(waktu="Pagi",nama="Mahasiswa"):
  print(f"Selamat {waktu}, {nama}")

selamat("Siang","Asep")
selamat("Sore")
selamat()
```

#4.	*Fungsi dengan args

```python
def jumlahkan_semua(*angka):
  return sum(angka)

hasil=jumlahkan_semua(3,5,7)
print(hasil)
```

#5.	**Fungsi dengan kwargs

```python
def info_barang(**data):
  for key, value in data.items():
    print(f"{key}:{value}")
info_barang(nama="Buku", harga=1500, stok=30)
```

#7.	Fungsi Lambda

```python
#a. menghitung kuadrat
kuadrat=lambda x:x**2
print("hasil kuadrat:",kuadrat(5))

#b. menghitung keliling lingkaran dengan jari-jari r
keliling=lambda r:2*3.14*r
print("Hasil keliling lingkaran:",keliling(4))
```

#8.	Fungsi Decorator

```python
def decorator_logger(fungsi):
  def wrapper():
    print("Memulai fungsi...")
    fungsi()
    print("Fungsi selesai dijalankan")
  return wrapper

@decorator_logger
def logger():
  print("Halo, selamat datang!")

logger()
```

#9.	Studi Kasus Nyata â€“ Program Nilai

```python
def input_nilai():
  nilai=[]
  jumlah=int(input(f"Masukkan jumlah nilai:"))
  for i in range(jumlah):
    n=(int(input(f"Masukkan nilai ke-{i+1}:")))
    nilai.append(n)
  return nilai

def rata_rata(nilai_list):
  if len(nilai_list)==0:
    return 0
  return sum(nilai_list)/len(nilai_list)

def status_kelulusan(rata):
  if rata>=60:
    return "LULUS"
  else:
    return "TIDAK LULUS"

nilai_akhir=lambda tugas, ujian: (tugas*0.4)+(ujian*0.6)

nilai_mahasiswa=input_nilai()
rata_rata_mahasiswa=rata_rata(nilai_mahasiswa)
status_lulus=status_kelulusan(rata_rata_mahasiswa)

print(f"Nilai mahasiswa: {nilai_mahasiswa}")
print(f"Rata-rata mahasiswa: {rata_rata_mahasiswa}")
print(f"Status kelulusan: {status_lulus}")

nilai_tugas = float(input("Masukkan nilai tugas: "))
nilai_ujian = float(input("Masukkan nilai ujian: "))
nilai_bobot=nilai_akhir(nilai_tugas,nilai_ujian)
status_akhir=status_kelulusan(nilai_bobot)

print(f"Nilai tugas: {nilai_tugas}")
print(f"Nilai ujian: {nilai_ujian}")
print(f"Nilai akhir: {nilai_bobot}")
print(f"Status kelulusan: {status_akhir}")
```
